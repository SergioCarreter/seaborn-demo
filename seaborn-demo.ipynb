```python
#Axes-level functions make self-contained plots

#Genero los subplots y guardo la figura en f y los axes en axs
#Como hay mas de 1 es una lista referenciable segun axs[n-1], con n=i·j

f, axs = plt.subplots(1, 2, figsize=(10, 6), gridspec_kw=dict(width_ratios=[4, 3]))
sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species", ax=axs[0])
sns.histplot(data=penguins, x="species", hue="species", shrink=.8, alpha=.8, legend=False, ax=axs[1])
f.tight_layout()
```

```python
tips=pd.read_csv('C:/Users/sergi/python/datasets/tips.csv')
```

```python
# Figure-level functions own their figure
# Nevertheless, it is possible to go beyond what the figure-level functions offer by accessing the matplotlib axes 
#               on the object that they return and adding other elements to the plot that way

g = sns.relplot(data=tips, x="total_bill", y="tip")
g.ax.axline(xy1=(10, 2), slope=.2, color="b", dashes=(5, 2))
```

```python
# Customizing plots from a figure-level function
g = sns.relplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", col="sex")
g.set_axis_labels("Flipper length (mm)", "Bill length (mm)")
```

```python
#Guardo en g la figura
g = sns.relplot(data=tips, x="total_bill", y="tip")
```

```python
g.ax
```

```python
g.ax.axline(xy1=(10, 2), slope=.2, color="b", dashes=(5, 2))
```

```python
#Guardo en g la figura
g = sns.relplot(data=tips, x="total_bill", y="tip")

# Dibujo una linea 'axline' en el eje 'ax' de la figura 'g'
g.ax.axline(xy1=(10, 2), slope=.2, color="b", dashes=(5, 2))
```

```python
# default output of matplotlib.pyplot.subplots() with one subplot:

f, ax = plt.subplots()
```

```python
# In Matplotlib: A figure with multiple columns will have the same overall size, 
# but the axes will be squeezed horizontally to fit in the space:

f, ax = plt.subplots(1, 2, sharey=True)
```

```python
# In Seaborn: a plot created by a figure-level function will be square. 
# To demonstrate that, let’s set up an empty plot by using FacetGrid directly. 
# This happens behind the scenes in functions like relplot(), displot(), or catplot():

g = sns.FacetGrid(penguins)
```

```python
# When additional columns are added, the figure itself will become wider, so that its subplots have the same size and shape:

g = sns.FacetGrid(penguins, col="sex")
```

```python
# And you can adjust the size and shape of each subplot 

g = sns.FacetGrid(penguins, col="sex", height=3.5, aspect=.75)
```

```python
# Combining multiple views on the data
# Two important plotting functions:
# jointplot() and pairplot() are figure-level functions create figures with multiple subplots by default
# But they use different objects to manage the figure: JointGrid and PairGrid, respectively
```

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species", height=9)
```

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species", kind="hist")
```

```python
sns.pairplot(data=penguins, hue="species")
```
